```python
def mostar_config(**kwargs):
    for k,v in kwargs.items():
        print(f'{k}: {v}')

mostar_config(usr='teo', ip='1,2,3,4',puerto='8080', altura=1.77)
```

```python
def log(mensaje,*args,**kwargs):
    nivel = kwargs.get('nivel', 'DEBUG')
    prefijo = kwargs.get('prefijo', '')
    # datos_formato = kwargs.get('')
    msj_formato=f'[{nivel}]: {prefijo} - {mensaje}'
    return msj_formato

log("Error de configuración", nivel='ERROR')
```

```python
def http_get(url, **kwargs):
    timeout = kwargs.get('timeout',5)
    headers = kwargs.get('headers', {})
    params = kwargs.get('params', {})

    return {
        'url' : url,
        'timeout': timeout,
        'headers' : headers,
        'params' : params
    }

http_get(
    'http://google.com',
    timeout=10
    params ={'page': 2}
    
)
```
