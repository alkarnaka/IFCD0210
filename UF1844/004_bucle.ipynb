# Bucles

```python
for i in range(10,1,-2):
    print(i)
```

```python
lista = [1,2,3,4,5,6,7,8,9,0]

for numero in lista:
    print(numero)
```

```python
for n in lista:
    if n % 2 == 0:
        print(n)
    else: 
        print('-')
```

# Bucles while

```python
contador = 0
while contador <= 10:
    print(contador)
    contador += 1
print('Fuera del bucle')

final = 10
control = 0
while True:
    print(f'hola {control}')
    if control == final:
        break
    control += 1
```
