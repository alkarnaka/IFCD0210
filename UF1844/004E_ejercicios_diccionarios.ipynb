1. Escriba un programa que solicite repetidamente al usuario el nombre y el precio de los productos. <br>
Almacene todos estos datos en un diccionario cuyas claves sean los nombres de los productos y cuyos valores sean los precios.<br>
Terminará al introducir un producto vacío.<br>
Se podrán introducir productos repetidos, pero solo se almacenarán una vez y se imprimirá el precio correspondiente o un mensaje si el producto no está en el diccionario.


````python
productos = {}
while True:
    nombre = input("Dime el nombre del Producto: ")
    if nombre == '':
        break

    precio = input("Dime el precio: ")
    # Por si se quiere guardar el primer valor introducido
    if not nombre in productos:
        productos[nombre] = float(precio)
    else:
        print(f'El producto "{nombre}" ya está en el diccionario')

from pprint import pprint
pprint(productos)
````


2. Usando el diccionario creado en el problema anterior, permita al usuario ingresar una cantidad en dólares e imprima todos los productos cuyo precio sea menor a esa cantidad.


````python
filtro = float(input("Dime el precio máximo: "))
for k,v in productos.items():
    if v < filtro:
        print(f'Producto: {k}, Precio: {v}')
````

8. Construya un diccionario con las cartas de la baraja, cree un juego de cartas simple que reparta tres cartas a dos jugadores. <br>
El jugador con la carta más alta gana. <br>
En caso de empate, se compara la segunda carta más alta y, si es necesario, la tercera más alta. <br>
Si las tres cartas tienen el mismo valor, el juego es un empate.


````python
import random
from pprint import pprint
baraja = []

for palo in ['oros','copas','espadas','bastos']:
    for i in range(1,13):
        carta ={palo:i}
        baraja.append(carta)

random.shuffle(baraja)

jugador1 = []
jugador2 = []

for i in range (3):
    jugador1.append(baraja.pop())
    jugador2.append(baraja.pop())

pprint(jugador1)
pprint(jugador2)

````
